```python
import pandas as pd
```

```python
df_raw = pd.read_csv('http://www.football-data.co.uk/mmz4281/1617/E0.csv')
df_raw.head()
```

```python
df = df_raw.loc[ : , 'Date':'FTAG']
df['Match'] = range(1, df.shape[0]+1)
df.Date = pd.to_datetime(df.Date)
df.head()
```

```python
df_melt = pd.melt(df, id_vars=['Match','Date','HomeTeam', 'AwayTeam'], 
                  value_vars=['FTHG', 'FTAG'], var_name='GoalPlace', value_name='Goals')
df_melt.sort_values(['Match']).head()
```

```python
df_melt['Team'] = df_melt.apply(
    lambda row: row['HomeTeam'] if row['GoalPlace'] == 'FTHG' else row['AwayTeam'], axis=1
)
df_melt.sort_values(['Match']).head()
del df_melt['AwayTeam']
del df_melt['GoalPlace']
```

```python
df_premier = df_melt.rename(columns={'HomeTeam': 'Location'})
df_premier = df_premier[['Match', 'Date', 'Location', 'Team', 'Goals']]
df_premier = df_premier.sort_values(['Match'])
df_premier = df_premier.reset_index()
del df_premier['index']
```
